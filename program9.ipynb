```python
import os
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate
from langchain_cohere import ChatCohere
from langchain_core.output_parsers import StrOutputParser
```

```python
from pydantic import BaseModel

class InstitutionDetails(BaseModel):
    founder: str
    founded: str
    branches: int
    employees: int
    summary: str
```

```python
prompt_template = """
Given the institution named '{institution_name}', extract the following details from Wikipedia:
1. Founder of the institution
2. When it was founded
3. Current branches of the institution
4. How many employees work in it
5. A 4-line brief summary of the institution """
```

```python
import getpass
import os

# Check if the COHERE_API_KEY environment variable is already set
if not os.environ.get("COHERE_API_KEY"):
    # If not set, prompt the user to enter their Cohere API key and set it as an environment variable
    os.environ["COHERE_API_KEY"] = getpass.getpass("Enter API key for Cohere: ")
```

```python
model = ChatCohere(model="command-r7b-12-2024")
prompt = PromptTemplate(input_variables=["institution_name"], template=prompt_template)
```

```python

chain = prompt | model
institution_name_input = input("Enter the institution name: ")

response = chain.invoke({"institution_name": institution_name_input})

print(f"Institution Name: {institution_name_input}\n")
print("Generated Response (Cohere Message object):")
print(response)

print("\nGenerated Response Content (string):")
print(response.content)
```
